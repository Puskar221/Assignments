based on their Score:

```python
def categorize_happiness(score):
    if score < 4:
        return 'Low'
    elif 4 <= score <= 6:
        return 'Medium'
    else:
        return 'High'
data['Happiness Category'] = data['score'].apply(categorize_happiness)
print(data[['Country name', 'score', 'Happiness Category']])
```

**Task - 3 - Outlier Detection:**

1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

```python
# Calculate Q1, Q3, and IQR for Score and GDP per Capita
score_q1, score_q3 = south_asia_data['score'].quantile([0.25, 0.75])
gdp_q1, gdp_q3 = south_asia_data['Log GDP per capita'].quantile([0.25, 0.75])

score_iqr = score_q3 - score_q1
gdp_iqr = gdp_q3 - gdp_q1

# Define lower and upper bounds for outliers
score_lower = score_q1 - 1.5 * score_iqr
score_upper = score_q3 + 1.5 * score_iqr
gdp_lower = gdp_q1 - 1.5 * gdp_iqr
gdp_upper = gdp_q3 + 1.5 * gdp_iqr

# Identify outliers
outliers = south_asia_data[
    (south_asia_data['score'] < score_lower) | (south_asia_data['score'] > score_upper) |
    (south_asia_data['Log GDP per capita'] < gdp_lower) | (south_asia_data['Log GDP per capita'] > gdp_upper)
]

# Display outliers
outliers[['Country name', 'score', 'Log GDP per capita']]
```

2. Define outliers using the 1.5 × IQR rule.

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = data['score'].quantile(0.25)
Q3 = data['score'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = data[(data['score'] < lower_bound) | (data['score'] > upper_bound)]

# Display the outliers
print(outliers[['Country name', 'score']])
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))

sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_data, color='blue', label='Normal')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', label='Outliers', s=100, marker='X')

plt.title('GDP per Capita vs Happiness Score with Outliers Highlighted')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')

plt.legend()

plt.show()
```
